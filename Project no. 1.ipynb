```python
import numpy as np
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv('data.csv')
```

```python
plt.figure(figsize=(10,12))
sns.set_style('darkgrid')
sns.scatterplot(x='MPG',y='Acceleration',data=df)
```

```python
plt.figure(figsize=(10,12))
sns.scatterplot(y='Acceleration',x='Horsepower',data=df)
```

```python
plt.figure(figsize=(10,12))
sns.scatterplot(x='Weight',y='Acceleration',data=df)
```

```python
plt.figure(figsize=(10,12))
sns.scatterplot(x='Weight',y="Horsepower",data=df)
```

```python
from sklearn.model_selection import train_test_split
x=df[['Weight','Acceleration']]
y=df['Horsepower']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.4, random_state=101)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
prediction=lm.predict(x_test)
```

```python
plt.scatter(y_test,prediction)
```

```python
sns.scatterplot(y_test,prediction)
```

```python
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, prediction))
print('Mean Square Error:', metrics.mean_squared_error(y_test, prediction))
print('Root Mean Square Error:', np.sqrt(metrics.mean_squared_error(y_test, prediction)))
print('R2 score:',metrics.r2_score(y_test,prediction))
```

With this my project is complete.It was a relatively short project where i have predicted the horsepower of a car based on weight and acceleration, as r2 score shows it is a moderate model to predict the horsepower but can be improved further without overfitting the data.
